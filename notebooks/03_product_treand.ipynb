```python
import pandas as pd
import matplotlib.pyplot as plt
from product_trends import product_trend
```

```python
df = pd.read_csv('../data/cleaned_data.csv')
product_name = 'WHITE HANGING HEART T-LIGHT HOLDER'
trend = product_trend(df, product_name)
```

```python
# สร้างกราฟ
plt.figure(figsize=(10,6))
trend.plot(kind='line', marker='o', color='b')
plt.title(f'Trend of Sales for {product_name}', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Quantity Sold', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
