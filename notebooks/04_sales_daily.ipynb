```python
import pandas as pd
import matplotlib.pyplot as plt
from daily_sales_stats import daily_sales

# อ่านข้อมูล
df = pd.read_csv('../data/cleaned_data.csv', parse_dates=['InvoiceDate'])

# ใช้ฟังก์ชัน daily_sales
daily_sales_stats = daily_sales(df)


# สร้างกราฟ Line Plot
plt.figure(figsize=(10, 6))
plt.plot(daily_sales_stats['Date'], daily_sales_stats['TotalPrice'], marker='o', color='b')  # ใช้ marker 'o' เพื่อแสดงจุด
plt.title('Daily Sales Trend', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)  # หมุนชื่อวันที่เพื่อให้อ่านง่าย
plt.grid(True)  # เปิดการแสดงตารางในกราฟ
plt.tight_layout()  # ปรับขนาดกราฟให้พอดีกับหน้าจอ
plt.show()  # แสดงกราฟ
```
