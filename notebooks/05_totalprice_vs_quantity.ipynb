```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from outlier_removal import filter_and_save
```

```python
# ระบุไฟล์และคอลัมน์ที่ต้องการกรอง
csv_path = '../data/cleaned_data.csv'
output_path = '../data/cleaned_data_filtered.csv'
columns_to_filter = ['Quantity', 'TotalPrice']

# เรียกใช้ฟังก์ชัน
filter_and_save(csv_path, output_path, columns_to_filter)
```

```python
# อ่านไฟล์ข้อมูล
df = pd.read_csv('../data/cleaned_data_filtered.csv')


# สร้างกราฟ Scatter Plot
plt.figure(figsize=(10,6))
plt.scatter(df['Quantity'], df['TotalPrice'], alpha=0.5, color='purple')
plt.title('Total Price vs Quantity Sold', fontsize=14)
plt.xlabel('Quantity Sold', fontsize=12)
plt.ylabel('Total Price (£)', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# สร้างกราฟ Scatter Plot พร้อมเส้นแนวโน้ม
plt.figure(figsize=(10,6))
sns.regplot(
    x='Quantity', 
    y='TotalPrice', 
    data=df, 
    scatter_kws={'alpha':0.5, 'color': 'purple'}, 
    line_kws={"color":"red", "linewidth":2}
)
plt.title('Total Price vs Quantity Sold with Trend Line', fontsize=14)
plt.xlabel('Quantity Sold', fontsize=12)
plt.ylabel('Total Price (£)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
